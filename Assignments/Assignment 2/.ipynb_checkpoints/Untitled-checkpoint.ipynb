```python
df
```

```python
df_group = df.groupby(" sa3_name16", as_index=False).sum()
```

```python
all_pop = {}

for i, row in df_group.iterrows():
    suburb = row[' sa3_name16']
    population = {}
    for name in df_group.columns[1:]:
        value = row[name]
        name = name.strip()
        name = name.replace('_', ' ')
        population[name] = value
    all_pop[suburb] = population

with open('adelaide.json', 'w') as f:
    f.write(json.dumps(all_pop))
```
